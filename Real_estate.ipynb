```python
new_colm_names = {'X1 transaction date' : 'x1tdate','X2 house age': 'x2hage',
                  'X3 distance to the nearest MRT station' : 'x3MRT',
                  'X4 number of convenience stores' : 'x4stores',
                  'X5 latitude' : 'x5lat',
                  'X6 longitude' : 'x6long','Y house price of unit area' : 'yprice'}
df = df.rename(columns = new_colm_names)
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
real_estate=df.astype(float)
```

```python
real_estate
```

```python
X=real_estate.iloc[:,:-2]
```

```python
y=real_estate.yprice
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model_estate=LinearRegression()
```

```python
model_estate.fit(X_train,y_train)
```

```python
print(model_estate)
```

```python
pred = model_estate.predict(X_test)
```

```python
print(pred)
```

```python
plt.scatter(y_test,pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.show()
```

```python
from sklearn import metrics
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```
